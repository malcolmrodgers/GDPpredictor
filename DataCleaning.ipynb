```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/malcolmrodgers/Documents/Coding/Python/GDPpredictor/worlddata2023.csv", sep=',')
df
```

```python
print(list(df.columns))
```

```python
df = df[['Country', 'GDP', 'Armed Forces size', 'Gross tertiary education enrollment (%)', 'Infant mortality', 'Life expectancy', 'Population', 'Urban_population', 'Latitude']]

df
```

```python
df.dtypes
```

```python
df = df.replace(',', '', regex=True)
df = df.dropna(axis=0)
df['GDP'] = df['GDP'].str.strip('$')
```

```python
cols = df.select_dtypes(object).columns
df[cols] = df[cols].apply(lambda x: x.str.strip('%'))
```

```python
df
```

```python
cols = list(df.loc[:, 'GDP':'Latitude'])
df[cols] = df[cols].astype(float)
df.dtypes
```
